```r
library(DESeq2)
```

# This is the less than .00001 data

## For paper analysis look below for "greater than .00001"

```r
data = read.table(file= "T5_filtered_less_000001.csv", header = T, sep = ',', check.names=FALSE, row.names=1)
data
```

```r
mapping = read.table(file = "Deseq2_T5_mapping.txt", header = T, sep="\t")
mapping
```

```r
dds = DESeqDataSetFromMatrix(countData = data, colData = mapping, design = ~ Treatment)
dds
```

```r
dds$Treatment = relevel(dds$Treatment, ref = "Let")
```

```r
dds = DESeq(dds, fitType="local")
dds
```

```r
result = results(dds)
res = result[order(result$padj),]
res
```

```r
write.table(res, file = "T5_filtered_less_000001_PvL_deseq.csv", sep = ",", quote = FALSE, row.names = TRUE, col.names = NA)
```

```r
LvCoL = results(dds, c("Treatment", "CoL", "Let"))
LvCoL = LvCoL[order(result$padj),]
write.table(LvCoL, file = "T5_filtered_less_000001_CoLvL_deseq.csv", sep = ",", quote = FALSE, row.names = TRUE, col.names = NA)
```
