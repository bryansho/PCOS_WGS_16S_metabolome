# T2 PvL (top 10) Log2Fold

### NOTE: Use top Log2Fold graphs not pvalue graphs

```r
sigtab = read.csv("T2_filtered_greater_00001_PvL_deseq.csv", row.names =1 )
sigtab = as.data.frame.matrix(sigtab)
row_sigtab = lapply(rownames(sigtab), function(x) {substring(x, 3)})
rownames(sigtab) = row_sigtab
sigtab
```

```r
summary(sigtab$log2FoldChange)
```

```r
rank_order = order(abs(sigtab$log2FoldChange), sigtab$log2FoldChange, decreasing = TRUE)
sigtab = sigtab[rank_order,]
sigtab
```

```r
x = 0
unwant = "uncultured"
threshold = data.frame()

for (i in row.names(sigtab)){
    if (grepl(unwant, i) == FALSE){
        threshold = rbind(threshold, sigtab[c(i),])
        x = x +1
    }
    if (x == 10){
        break
    }
}

threshold
```

```r
threshold = threshold[order(row.names(threshold)),]
threshold
```

```r
pdf('T2_WSG_PvsL_top10log_DESEQ_v2.pdf', width=6, height=8)
theme_set(theme_bw())
ggplot(threshold, aes(y= factor(row.names(threshold),levels= rev(levels(factor(row.names(threshold))))), x=threshold$log2FoldChange)) + 
  geom_vline(xintercept = 0.0, color = "gray", size = 0.5) +
  geom_point(size=2.5) + 
  ggtitle("WSG P/L T2 (Top 10 Log2fold)") +
  xlab("log2FoldChange") +
  ylab("Bacterial Species") + 
  scale_x_continuous(minor_breaks=NULL) +
  theme(text= element_text(size=10), axis.text.x = element_text(angle = -90, hjust = 0, vjust=0.5),
        legend.position = 'none', plot.title = element_text(hjust = 0.5), axis.text=element_text(size=12)) +
  coord_fixed(ratio=.3) 
dev.off()
```
