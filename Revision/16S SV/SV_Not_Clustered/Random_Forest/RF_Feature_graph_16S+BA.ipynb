```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("RF_top_features.csv")
df
```

```python
split_group = df.groupby("Comparison")
data = [split_group.get_group(x) for x in split_group.groups]
order = [1,3,0,2]
ordered_data = []
for i in order:
    display(data[i])
    ordered_data.append(data[i])
```

```python
data
```

```python
for i, graph in enumerate(ordered_data):
    print(i//2)
    print(i%2)
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(20, 15))
fig.subplots_adjust(hspace=.3,wspace=0.65)
Titles = ["BA+16s P/LET Week 2", "BA+16s P/LET Week 5", "BA+16s LETch/LET Week 2", "BA+16s LETch/LET Week 5"]

for i, graph in enumerate(ordered_data):
    graph.plot(kind='barh', x ="Feature", legend=False,ax=axs[i//2][i%2])
    axs[i//2][i%2].set_title(Titles[i])
    axs[i//2][i%2].invert_yaxis()
    axs[i//2][i%2].set_ylabel("")
    axs[i//2][i%2].set_xlim([0, 0.22])
    
fig.text(0.45, 0.087, 'Importance', ha='center', size = 20)
fig.text(-.045, 0.5, 'Features', va='center', rotation='vertical', size = 20)   
fig.align_ylabels(axs[:, 0])
fig.align_ylabels(axs[:, 1])
plt.savefig("RF_Features_importance.pdf",format='pdf', dpi=650,bbox_inches = "tight")
```
