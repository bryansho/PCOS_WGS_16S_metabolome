## Importing all files and making sure that they are compatibale with the analysis (Pre-Analysis)

```r
#Make sure the data has the sample names on rows if not just transform the data with t()
abund_table<-read.csv("table_cluster.tsv",sep ='\t',row.names=1,check.names=FALSE)
meta_table<-read.csv("mapping_file_cohousing2018.tsv",sep='\t',row.names=1,check.names=FALSE)
#Transpose the data to have sample names on rows
abund_table<-t(abund_table)

meta_table<-data.frame(meta_table)
#Filter out samples not present in meta_table
abund_table<-abund_table[rownames(abund_table) %in% rownames(meta_table),]

#Now load the taxonomy
OTU_taxonomy<-read.csv("taxonomy_phyloseq.csv",row.names=1,check.names=FALSE)
abund_table
meta_table
```

```r
#Convert the data to phyloseq format
OTU = otu_table(as.matrix(abund_table), taxa_are_rows = FALSE)
TAX = tax_table(as.matrix(OTU_taxonomy))
SAM = sample_data(meta_table)
phy_main<-merge_phyloseq(phyloseq(OTU, TAX),SAM)
```

```r
#nsamples() shows the number of samples
print ('Number of Samples in our Biom Table')
nsamples(phy_main)

#ntaxa() shows number of unique OTUs there are 
print ('Number of OTUs we have.')
ntaxa(phy_main)

#sample_sums() shows summary statistics of the sampling depth (sequences per sample)
print ('Summary stats of sequencing depth.')
depths <- sample_sums(phy_main)
summary(depths)

#sort() sorts the phyloseq object. use this to find the sample with the lowest sequences
print("Sample with the lowest amount of sequences")
head(sort(sample_sums(phy_main)),1)

#rank_names() shows the taxa ranks of phylo obj
print("Taxa ranks")
rank_names(phy_main)

#Name of phylo obj shows the components of the obj
print("components of phylo object")
phy_main

#Showing the actual taxa of phy obj
print("First 10 rep taxa from phy obj")
taxtable=tax_table(phy_main)
taxtable[1:10]
```

```r
phy.pt = subset_samples(phy_main, Week != '0') # removing time 0
phy.pt.time2 = subset_samples(phy.pt, Week != '1' & Week!='3' & Week!='4' & Week!='5') # removing time 0
phy.pt.time5 = subset_samples(phy.pt, Week != '1' & Week!='3' & Week!='4' & Week!='2') # removing time 0
phy.pt.PL.time2 = subset_samples(phy.pt.time2, Category != 'Co-P' & Category != 'Co-L')
phy.pt.PL.time5 = subset_samples(phy.pt.time5, Category != 'Co-P' & Category != 'Co-L')
phy.pt.LCoL.time2 = subset_samples(phy.pt.time2, Category != 'Placebo' & Category != 'Co-P')
phy.pt.LCoL.time5 = subset_samples(phy.pt.time5, Category != 'Placebo' & Category != 'Co-P')
phy.pt.PL = subset_samples(phy.pt, Category != 'Co-P' & Category != 'Co-L')
phy.pt.CoLCoP = subset_samples(phy.pt, Category != 'Placebo' & Category != 'Letrozole')
phy.pt.LCoL = subset_samples(phy.pt, Category != 'Placebo' & Category != 'Co-P')
phy.pt.LCoP = subset_samples(phy.pt, Category != 'Placebo' & Category != 'Co-L')
phy.pt.PCoL = subset_samples(phy.pt, Category != 'Letrozole' & Category != 'Co-P')
phy.pt.PCoP = subset_samples(phy.pt, Category != 'Letrozole' & Category != 'Co-L')
```

```r
deseq2.PL <- phyloseq_to_deseq2(phy.pt.PL, ~ Category)
deseq2.ColCoP <- phyloseq_to_deseq2(phy.pt.CoLCoP, ~ Category)
deseq2.LCoL <- phyloseq_to_deseq2(phy.pt.LCoL, ~ Category)
deseq2.LCoP <- phyloseq_to_deseq2(phy.pt.LCoP, ~ Category)
deseq2.PCoL <- phyloseq_to_deseq2(phy.pt.PCoL, ~ Category)
deseq2.PCoP <- phyloseq_to_deseq2(phy.pt.PCoP, ~ Category)
deseq2.PL.time2 <- phyloseq_to_deseq2(phy.pt.PL.time2, ~ Category)
deseq2.PL.time5 <- phyloseq_to_deseq2(phy.pt.PL.time5, ~ Category)
deseq2.LCoL.time2 <- phyloseq_to_deseq2(phy.pt.LCoL.time2, ~ Category)
deseq2.LCoL.time5 <- phyloseq_to_deseq2(phy.pt.LCoL.time5, ~ Category)
```

```r
#checking order of deseq2 objects
unique(sample_data(phy.pt)$Category)
unique(sample_data(phy.pt)$Week)
unique(sample_data(phy.pt.CoLCoP)$Category)
unique(sample_data(phy.pt.CoLCoP)$Week)
unique(sample_data(phy.pt.PL)$Category)
unique(sample_data(phy.pt.PL)$Week)
unique(sample_data(phy.pt.LCoL)$Category)
unique(sample_data(phy.pt.LCoL)$Week)
unique(sample_data(phy.pt.LCoP)$Category)
unique(sample_data(phy.pt.LCoP)$Week)
unique(sample_data(phy.pt.PCoL)$Category)
unique(sample_data(phy.pt.PCoL)$Week)
unique(sample_data(phy.pt.PCoP)$Category)
unique(sample_data(phy.pt.PCoP)$Week)
unique(sample_data(phy.pt.PL.time2)$Category)
unique(sample_data(phy.pt.PL.time2)$Week)
unique(sample_data(phy.pt.PL.time5)$Category)
unique(sample_data(phy.pt.PL.time5)$Week)
unique(sample_data(phy.pt.LCoL.time2)$Category)
unique(sample_data(phy.pt.LCoL.time2)$Week)
unique(sample_data(phy.pt.LCoL.time5)$Category)
unique(sample_data(phy.pt.LCoL.time5)$Week)
#deseq2$Category <- relevel(deseq2$Category, ref = "Placebo")
```

# Placebo Vs Letrozole T2

```r
# calculate geometric means prior to estimate size factors
gm_mean = function(x, na.rm=TRUE){
  exp(sum(log(x[x > 0]), na.rm=na.rm) / length(x))
}
geoMeans = apply(counts(deseq2.PL.time2), 1, gm_mean)
deseq2.PL.time2 = estimateSizeFactors(deseq2.PL.time2, geoMeans = geoMeans)
deseq2.PL.time2 = DESeq(deseq2.PL.time2, fitType="local")

res = results(deseq2.PL.time2)
res = res[order(res$log2FoldChange, na.last=NA), ]
alpha = 1
#sigtab = res[(res$padj < alpha), ]
sigtab = res
sigtab = cbind(as(sigtab, "data.frame"), as(tax_table(phy.pt.PL.time2)[rownames(sigtab), ], "matrix"))
sigtab
head(sigtab, contrasts=list("Placebo","Letrozole"))

res
```
